```
import gensim, wikipedia
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS
import wikipedia
from gensim.summarization.summarizer import summarize
from gensim.summarization.keywords import keywords
import nltk
nltk.download('rslp')
nltk.download('punkt')
nltk.download('stopwords')
from nltk import word_tokenize, RSLPStemmer
from nltk.corpus import stopwords
```

```
# wikipedia.set_lang('pt')
wikisearch = wikipedia.page('Usina Hidrelétrica de Itaipu')
wikicontent = wikisearch.content
print('')
print(wikicontent)
```

```
print("\n##### Resumos ####")
summ_per = summarize(wikicontent, ratio=0.01)
print(summ_per)

summ_words = summarize(wikicontent, word_count=140)
print(summ_words)
```

```
key_words = keywords(wikicontent, ratio=0.05, pos_filter=('NN'), words = 5)
print(key_words)
```

```
wikipedia.set_lang('pt')
# wikisearch2 = wikipedia.page('python')
wikisearch2 = wikipedia.page('Usina Hidrelétrica de Itaipu')
wikicontent2 = wikisearch2.content
# print('')
# print(wikicontent2)
summ_words2 = summarize(wikicontent2, word_count=250)
print(summ_words2)

stopwords = stopwords.words('portuguese')
tokens_singles = nltk.tokenize.word_tokenize(wikicontent2.lower())
remove = [',', '.', '\'', '/', '-', '..', '!', '[', ']', 's', '?', ':', '%', '&', '*', '#', 'f', 'r', 'ha', 'heh', '...']
texto = []

for p in tokens_singles:
    if p not in stopwords and p not in remove:
        texto.append(p)

texto = ' '.join(texto)
key_words = keywords(texto, ratio=0.05, pos_filter=('NN'), words=5)
print(key_words)

# stemming
# rslp = RSLPStemmer()
# tokens_pt = word_tokenize(summ_words2)
# stems_rslp = []
# for t in tokens_pt:
#   stems_rslp.append(rslp.stem(t))

# print(stems_rslp)
# key_words2 = keywords(' '.join(stems_rslp), words = 5)
# print(key_words2)
```

```
text = 'all your base are belong to us all of your base base base'

def generate_wordcloud(text):
  wordcloud = WordCloud(width = 800, height = 400,  relative_scaling = 1.0,   stopwords = {'to', 'of', 'us'}).generate(text)
  plt.axis('off')
  plt.imshow(wordcloud)
  plt.savefig('WordCloud.png')
  plt.show()
```

```
generate_wordcloud(text)
```
